```python
with zipfile.ZipFile('ZippedData/im.db.zip', mode='r') as archive:
    with archive.open('im.db', mode='r') as db:
        conn = sqlite3.connect(db.read())
        query = '''
                    SELECT *
                    FROM sqlite_master
                    ;
                    '''
        df = pd.read_sql(query, conn)

df
```
